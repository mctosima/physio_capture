```python
import datetime as dt
```

```python
# get current hour and minute
now = dt.datetime.now()
print(now)
now_hm = now.strftime("%H:%M")
print(now_hm)
```

```python
# add 90 secs from now_hm
now_hm = dt.datetime.strptime(now_hm, "%H:%M")
now_hm = now_hm + dt.timedelta(seconds=60)
now_hm = now_hm.strftime("%H:%M:%S")
print(now_hm)
```
